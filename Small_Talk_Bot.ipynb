```python
#creating the initial project 

from rasa.cli.scaffold import create_initial_project
```

```python
# creating a working directory and storing the project 

project = "small_talk"

create_initial_project(project)
```

```python
# move into project directory and cheking the files
os.chdir(project)
print(os.listdir("."))
```

```python
#creating some variables to pass to our chatbot model later

config = "config.yml"
training_files = "data/"
domain = "domain.yml"
output = "models/"          #storing our model
print(config, training_files, domain, output)
```

```python
model_path = rasa.train(domain, config, [training_files], output) #training files are passed here as a list
print(model_path)
```

```python
#let's check it out !

from rasa.jupyter import chat

#declaring the endpoint
endpoints = 'endpoints.yml'

chat(model_path, endpoints)
```
